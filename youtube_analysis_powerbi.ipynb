```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Read the Excel file
songs_data = pd.read_excel("C:/Users/NTC/Downloads/songs.xlsx")

# Display the first few rows of the dataset
songs_data.head()
```

```python
songs_data.drop(columns=['favoriteCount'] , inplace=True)
```

```python
# Get a summary of the dataset
print(songs_data.info())

# Check for missing values
print(songs_data.isnull().sum())
```

```python
# Example: Fill missing values with the median or drop rows
#songs_data = songs_data.fillna(songs_data.median())
# Or
songs_data = songs_data.dropna()
```

```python
# Calculate basic statistics
print(songs_data.describe())
```

```python


# Select numeric columns only
numeric_columns = songs_data.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr = numeric_columns.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
